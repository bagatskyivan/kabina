### 2. Prove:  
$$\frac{d}{dx} \left(\ln(\sin(x))\right) = \cot(x)$$

**Solution:**

1. Start by differentiating the natural logarithm:  
   $$\frac{d}{dx} \left(\ln(\sin(x))\right) = \frac{1}{\sin(x)} \cdot \frac{d}{dx}(\sin(x))$$

2. Differentiate $\sin(x)$:  
   $$\frac{d}{dx}(\sin(x)) = \cos(x)$$

3. Substitute into the expression:  
   $$\frac{1}{\sin(x)} \cdot \cos(x) = \frac{\cos(x)}{\sin(x)}$$

4. Recognize the definition of the cotangent function:  
   $$\frac{\cos(x)}{\sin(x)} = \cot(x)$$

**Conclusion:**  
$$\frac{d}{dx} \left(\ln(\sin(x))\right) = \cot(x)$$

### 3. Verify that $f''(x) = -f(x)$ for $f(x) = \cos(x)$.

**Solution:**

1. Start with $f(x) = \cos(x)$.

2. Find the first derivative:
   $$f'(x) = -\sin(x)$$

3. Differentiate again to find the second derivative:
   $$f''(x) = -\cos(x)$$

4. Observe that:
   $$f''(x) = -f(x)$$

**Conclusion:**  
$$f''(x) = -f(x) \quad \text{is verified.}$$

### 4. Using de l'Hôpital's Rule, find the improper limits:



#### (a) $$\lim_{x \to 0} \frac{\sin(x)}{x}$$

**Solution:**

1. The limit is of the indeterminate form \(\frac{0}{0}\). Apply de l'Hôpital's Rule:
   $$\lim_{x \to 0} \frac{\sin(x)}{x} = \lim_{x \to 0} \frac{\frac{d}{dx}(\sin(x))}{\frac{d}{dx}(x)}.$$

2. Differentiate the numerator and denominator:
   $$\frac{d}{dx}(\sin(x)) = \cos(x), \quad \frac{d}{dx}(x) = 1.$$

3. Substitute the derivatives:
   $$\lim_{x \to 0} \frac{\cos(x)}{1} = \cos(0) = 1.$$

**Conclusion:**
$$\lim_{x \to 0} \frac{\sin(x)}{x} = 1.$$

---

#### (b) $$\lim_{x \to \infty} \frac{\ln(x)}{x}$$

**Solution:**

1. The limit is of the indeterminate form $(\frac{\infty}{\infty})$. Apply de l'Hôpital's Rule:
   $$\lim_{x \to \infty} \frac{\ln(x)}{x} = \lim_{x \to \infty} \frac{\frac{d}{dx}(\ln(x))}{\frac{d}{dx}(x)}.$$

2. Differentiate the numerator and denominator:
   $$\frac{d}{dx}(\ln(x)) = \frac{1}{x}, \quad \frac{d}{dx}(x) = 1.$$

3. Substitute the derivatives:
   $$\lim_{x \to \infty} \frac{\frac{1}{x}}{1} = \lim_{x \to \infty} \frac{1}{x} = 0.$$

**Conclusion:**
$$\lim_{x \to \infty} \frac{\ln(x)}{x} = 0.$$

---

#### (c) $$\lim_{x \to \infty} \frac{\exp(x)}{x}$$

**Solution:**

1. The limit is of the indeterminate form $(\frac{\infty}{\infty})$. Apply de l'Hôpital's Rule:
   $$\lim_{x \to \infty} \frac{\exp(x)}{x} = \lim_{x \to \infty} \frac{\frac{d}{dx}(\exp(x))}{\frac{d}{dx}(x)}.$$

2. Differentiate the numerator and denominator:
   $$\frac{d}{dx}(\exp(x)) = \exp(x), \quad \frac{d}{dx}(x) = 1.$$

3. Substitute the derivatives:
   $$\lim_{x \to \infty} \frac{\exp(x)}{1} = \lim_{x \to \infty} \exp(x).$$

4. Since \(\exp(x)\) grows unbounded as $(x \to \infty)$:
   $$\lim_{x \to \infty} \exp(x) = \infty.$$

**Conclusion:**
$$\lim_{x \to \infty} \frac{\exp(x)}{x} = \infty.$$

### 5. Position, Velocity, and Acceleration of a Particle

The position of the particle is given by:
$$x(t) = 3t^2 - 6t + 1.$$

#### (a) Velocity, \(V(t)\):

1. The velocity is the first derivative of the position function:
   $$V(t) = x'(t).$$

2. Differentiate \(x(t)\):
   $$x'(t) = \frac{d}{dt}\left(3t^2 - 6t + 1\right).$$

3. Compute the derivative:
   $$x'(t) = 6t - 6.$$

So, the velocity is:
$$V(t) = 6t - 6.$$

---

#### (b) Acceleration, \(a(t)\):

1. The acceleration is the first derivative of the velocity (or the second derivative of the position):
   $$a(t) = V'(t) = x''(t).$$

2. Differentiate \(V(t)\):
   $$V'(t) = \frac{d}{dt}(6t - 6).$$

3. Compute the derivative:
   $$V'(t) = 6.$$

So, the acceleration is:
$$a(t) = 6.$$

---

#### (c) Velocity and Acceleration at \(t = 2\):

1. Substitute \(t = 2\) into the velocity equation:
   $$V(2) = 6(2) - 6 = 12 - 6 = 6.$$

2. Substitute \(t = 2\) into the acceleration equation:
   $$a(2) = 6.$$

---

**Conclusion:**  
At \(t = 2\):  
- Velocity, \(V(2) = 6\),  
- Acceleration, \(a(2) = 6\).
